# Import DATA

```python
from urllib.request import urlretrieve
urlretrieve('https://raw.githubusercontent.com/rishabhnmishra/Python_Diwali_Sales_Analysis/main/Diwali%20Sales%20Data.csv','diwali_sales_data.csv')
```

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Data Cleaning Process

```python
sales_df=pd.read_csv('diwali_sales_data.csv',encoding='unicode_escape')
```

# Check Null/Empty Value In Data

```python
pd.isnull(sales_df).sum()
```

# Removing Two Columns with null values

```python
sales_df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

# Now Removing Amount Rows with Null Values

```python
sales_df.dropna(inplace=True)
sales_df['Amount'].dtypes
```

# Now changing Amount Column dtype from float dtype to int dtype 

```python
sales_df['Amount']=sales_df['Amount'].astype('int')
sales_df['Amount'].dtypes
```

# Exploratory Data Analysis

# Gender Column:

```python
ax=sns.countplot(data=sales_df,x='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grp_gen=sales_df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=grp_gen,x='Gender',y='Amount',hue='Gender')
```

#### *From above graph we can see that most of the buyers are females and even the purchasing power of females are greater than men*

# Age Group Column:

```python
sns.set(rc={'figure.figsize':(10,6)})
ax=sns.countplot(data=sales_df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_gp=sales_df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=age_gp,x='Age Group',y='Amount')
```

#### *From above graph we can see that most of the buyers are of age group between 26-35 yrs female*

# State Column:

```python
st_or=sales_df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=st_or,x='State',y='Orders')
```

```python
st_or=sales_df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=st_or,x='State',y='Amount')
```

#### *From above graphs we can see that unexpectedly most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively but total sales/Amount is from Uttar Pradesh,Maharashtra and then Karnataka

# Marital Status Column:

```python
sns.set(rc={'figure.figsize':(6,5)})
ax=sns.countplot(data=sales_df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
mt=sales_df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=mt,x='Marital_Status',y='Amount')
```

#### *From above graphs we can see that most of the buyers are married(Women) and they have high purchasing power*

# Occupation Column:

```python
sns.set(rc={'figure.figsize':(23,10)})
ax=sns.countplot(data=sales_df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occ=sales_df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=occ,x='Occupation',y='Amount')
```

#### *From above graphs we can see that most of the buyers are working in IT,Aviation and HealthCare Sector*

# Product Category Columns:

```python
sns.set(rc={'figure.figsize':(30,15)})
ax=sns.countplot(data=sales_df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
pro_ct=sales_df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(40,15)})
sns.barplot(data=pro_ct,x='Product_Category',y='Amount')
```

#### *From above graphs we can see that most of the sold products are from Food, Clothings and Electronics Category*

# OverAll Conclusions:
#### *Married women age group 26-35yrs from Uttar Pradesh,Maharashtra and Karnataka working in IT,HealthCare and Aviation are most likely to buy products from food clothing and electronics category*
